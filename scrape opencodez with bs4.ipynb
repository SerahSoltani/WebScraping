```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.opencodez.com/"
headers = {
    "User-Agent": "Mozilla/5.0"
}
response = requests.get(url, headers=headers)
response
```

```python
bs  = BeautifulSoup(response.content, "html.parser")
bs
```

```python
number_of_pages = int(bs.find_all("a", attrs={"class": "inactive"})[-1]['href'].split("/")[-1])
number_of_pages
```

```python
title = []
description = []
img_src_link = []
author = []
month_day = []
year_list = []
post_link_list = []

number_of_pages = 19

for p in range(1, number_of_pages + 1):
    url = f"https://www.opencodez.com/page/{p}"
    response = requests.get(url)
    bs = BeautifulSoup(response.content, "html.parser")

    latest_post = bs.find("div", attrs={'class': 'latest-section'})
    if latest_post is None:
        print(f"'latest-section' not found in {url}")
        continue

    posts = latest_post.find_all("article", class_="post")
    print(f"Found {len(posts)} posts on page {p}.")

    for post in posts:
        # Title
        t = post.find("h2", class_="title")
        title.append(t.get_text(strip=True) if t else "")

        # Description
        desc = post.find("div", class_="post-content")
        p_tag = desc.find("p") if desc else None
        description.append(p_tag.get_text(strip=True) if p_tag else "")

        # Image link
        img_div = post.find("div", class_="featured-thumbnail")
        img_tag = img_div.find("img") if img_div else None
        img_src_link.append(img_tag['src'] if img_tag and img_tag.has_attr('src') else "")

        # Author
        info = post.find("div", class_="post-info")
        author_tag = info.find("a") if info else None
        author.append(author_tag.get_text(strip=True) if author_tag else "")

        # Date
        date_tag = post.find("span", class_="thetime")
        if date_tag:
            date_parts = date_tag.get_text(strip=True).split(",")
            month_day.append(date_parts[0])
            year_list.append(date_parts[1].strip() if len(date_parts) > 1 else "")
        else:
            month_day.append("")
            year_list.append("")

        # Post link
        link_tag = t.find("a") if t else None
        post_link_list.append(link_tag['href'] if link_tag and link_tag.has_attr('href') else "")

    print(f"Scraping page number {p} is finished.")
```

### create Pandas Dataframe

```python
import pandas as pd

df = pd.DataFrame(list(zip(
    title,
    description,
    author,
    year_list,
    month_day,
    post_link_list,
    img_src_link
)), columns=['Title', 'Description', 'Author', 'Year', 'Month', 'Link', 'Imagelink'])

df.head()
```

```python
df.to_csv("opencodez_posts.csv", index=False)
```
